1X1 1. 1  
2X2 2. 1 3 = 4   
3X3 4. 1 5 9 = 15   
4X4 6. 1 7 13 = 21  
5X5 8. 1 9 17 25 = 52  
nXn n>1, 2(n-1). 

```python
def solution(n):
  res = 1
  cnt = n-1
  while n > 1:
    res += res + 2*(n-1)
    cnt -= 1
    if cnt == 0:
      break

  print(res)
```

```python
solution(1)
solution(2)
solution(3)
solution(4)
solution(5)
```
