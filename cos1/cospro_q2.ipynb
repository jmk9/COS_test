# 문제풀이

```python
def func_a(string, length):
  padZero = ''
  padSize = length - len(string)
  for i in range(padSize):
    padZero += '0'

  return padZero + string
```

```python
def solution(binA, binB):
  max_length = max(len(binA), len(binB))
  binA = func_a(binA, max_length)
  binB = func_a(binB, max_length)

  print(binA)
  print(binB)

  hamming_dis = 0
  for i in range(max_length):
    if binA[i] != binB[i]:
      hamming_dis += 1

  return hamming_dis
```
